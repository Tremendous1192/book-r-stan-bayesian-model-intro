# 5-3-状態空間モデルによる予測と補間

```python
# -*- coding: utf-8 -*-
"""
Local level state-space model (forecasting & interpolation) in NumPyro
- Data I/O: pandas
- Bayesian inference: NumPyro (NUTS)
- Model visualization: numpyro.contrib.render.render_model
- Posterior/posterior predictive visualization: ArviZ (hdi_prob)
- Time-series plots: matplotlib (labels in English)
- Constraints:
  * Use print() to display results
  * Do NOT use az.from_numpyro(..., observed_data=...) nor az.from_numpyro(mcmc=...)
  * Do NOT use az.plot_density(..., kind=...)
  * When plotting PPC, pass group="posterior"
  * Do NOT pass group=... to az.plot_forest(...)
"""

import os
import numpy as np
import pandas as pd
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az

import numpyro
import numpyro.distributions as dist
from numpyro.infer import MCMC, NUTS, Predictive
from numpyro.contrib.render import render_model

# --------------------------------------------------------------------------------------
# 0) Utility: plotting helpers (time-series ribbons), HDI via ArviZ
# --------------------------------------------------------------------------------------

def plot_ts_with_hdi(dates, draws_2d, obs_dates=None, obs_values=None,
                     title="Result", ylabel="Value", xlabel="Date",
                     hdi_prob=0.95, legend_loc="best"):
    """
    Plot median + HDI ribbon for time series.
    - draws_2d: shape (n_draws, T)
    """
    # Median & HDI
    median = np.median(draws_2d, axis=0)
    hdi = az.hdi(draws_2d, hdi_prob=hdi_prob)  # shape (T, 2)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(dates, median, lw=2, label=f"Median ({hdi_prob*100:.0f}% HDI)")
    ax.fill_between(dates, hdi[:, 0], hdi[:, 1], alpha=0.3, label="HDI ribbon")

    if (obs_dates is not None) and (obs_values is not None):
        ax.scatter(obs_dates, obs_values, s=18, alpha=0.8, label="Observed")

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(loc=legend_loc)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


# --------------------------------------------------------------------------------------
# 1) Data loading (pandas)
# --------------------------------------------------------------------------------------

# Read full data (no missing)
sales_df_all = pd.read_csv("5-2-1-sales-ts-1.csv")
sales_df_all["date"] = pd.to_datetime(sales_df_all["date"])
y_all = sales_df_all["sales"].to_numpy()
T_all = len(y_all)

# For forecasting plot dates (T + pred_term)
pred_term = 20
date_plot = pd.date_range(start="2010-01-01", periods=T_all + pred_term, freq="D")

# Read data with missing values
sales_df_NA = pd.read_csv("5-3-1-sales-ts-1-NA.csv")
sales_df_NA["date"] = pd.to_datetime(sales_df_NA["date"])
y_NA = sales_df_NA["sales"].to_numpy()
T_NA = len(y_NA)

# Observed-only subset for interpolation model
sales_df_omit_NA = sales_df_NA.dropna(subset=["sales"]).copy()
y_obs_only = sales_df_omit_NA["sales"].to_numpy()
# R -> Python index shift: R is 1-based; Python is 0-based
obs_idx = np.where(~np.isnan(y_NA))[0].astype(int)  # 0-based indices

# Quick checks via print()
print("Full data length (T_all):", T_all)
print("Missing-data total length (T_NA):", T_NA)
print("Number of observed points in NA data (len_obs):", len(y_obs_only))
print("First 3 observed indices (0-based):", obs_idx[:3])


# --------------------------------------------------------------------------------------
# 2) NumPyro models
# --------------------------------------------------------------------------------------

def local_level_model_forecast(y, pred_term):
    """
    Local level model with future state forecasting.
    State:
        mu[0] ~ Normal(0, 10)
        mu[t] = mu[t-1] + s_w * eta[t-1],   eta ~ Normal(0, 1)
    Observation:
        y[t] ~ Normal(mu[t], s_v)
    Generated:
        mu_pred: length T + pred_term (states incl. forecasted)
    """
    T = y.shape[0]
    s_w = numpyro.sample("s_w", dist.Exponential(1.0))  # >=0
    s_v = numpyro.sample("s_v", dist.Exponential(1.0))  # >=0

    mu0 = numpyro.sample("mu0", dist.Normal(0.0, 10.0))
    eta = numpyro.sample("eta", dist.Normal(0.0, 1.0).expand([T - 1]).to_event(1))

    # Build latent states deterministically from innovations
    mu_increments = s_w * eta
    mu = jnp.concatenate([jnp.array([mu0]),
                          mu0 + jnp.cumsum(mu_increments)])
    numpyro.deterministic("mu", mu)

    # Likelihood
    numpyro.sample("y", dist.Normal(mu, s_v), obs=y)

    # Forecast beyond T
    eta_pred = numpyro.sample("eta_pred", dist.Normal(0.0, 1.0).expand([pred_term]).to_event(1))
    mu_future = mu[-1] + jnp.cumsum(s_w * eta_pred)
    mu_pred = jnp.concatenate([mu, mu_future])
    numpyro.deterministic("mu_pred", mu_pred)


def local_level_model_interpolation(y_obs, obs_idx, T):
    """
    Local level model with missing observations handled via indexing.
    Only y at obs_idx are observed.
    """
    s_w = numpyro.sample("s_w", dist.Exponential(1.0))
    s_v = numpyro.sample("s_v", dist.Exponential(1.0))

    mu0 = numpyro.sample("mu0", dist.Normal(0.0, 10.0))
    if T == 1:
        mu = jnp.array([mu0])
    else:
        eta = numpyro.sample("eta", dist.Normal(0.0, 1.0).expand([T - 1]).to_event(1))
        mu = jnp.concatenate([jnp.array([mu0]),
                              mu0 + jnp.cumsum(s_w * eta)])
    numpyro.deterministic("mu", mu)

    # Observe only at obs_idx
    numpyro.sample("y_obs", dist.Normal(mu[obs_idx], s_v), obs=y_obs)


def local_level_model_pred_interval(y_obs, obs_idx, T):
    """
    Same as interpolation, but also generate y_pred over entire time range for PPC.
    """
    s_w = numpyro.sample("s_w", dist.Exponential(1.0))
    s_v = numpyro.sample("s_v", dist.Exponential(1.0))

    mu0 = numpyro.sample("mu0", dist.Normal(0.0, 10.0))
    if T == 1:
        mu = jnp.array([mu0])
    else:
        eta = numpyro.sample("eta", dist.Normal(0.0, 1.0).expand([T - 1]).to_event(1))
        mu = jnp.concatenate([jnp.array([mu0]),
                              mu0 + jnp.cumsum(s_w * eta)])
    numpyro.deterministic("mu", mu)

    # Observe only at obs_idx
    numpyro.sample("y_obs", dist.Normal(mu[obs_idx], s_v), obs=y_obs)

    # Posterior predictive for all time points
    numpyro.sample("y_pred", dist.Normal(mu, s_v))


# --------------------------------------------------------------------------------------
# 3) Run inference (NUTS) & visualize model graphs (NumPyro built-in)
# --------------------------------------------------------------------------------------

rng_key = jax.random.PRNGKey(1)
num_warmup, num_samples, num_chains = 1000, 2000, 4

# 3.1 Forecasting (mu_pred)
nuts1 = NUTS(local_level_model_forecast)
mcmc1 = MCMC(nuts1, num_warmup=num_warmup, num_samples=num_samples, num_chains=num_chains, progress_bar=False)
mcmc1.run(rng_key, y=jnp.asarray(y_all), pred_term=pred_term)
mcmc1.print_summary()  # NumPyro's own summary (prints to stdout)
# Also print an ArviZ summary via print()
#idata1 = az.from_numpyro(
#    posterior=mcmc1.get_samples(group_by_chain=True),
#    sample_stats=mcmc1.get_extra_fields(group_by_chain=True)
#)
idata1 = az.from_numpyro(posterior=mcmc1)
print("\n[Forecast model] ArviZ summary (s_w, s_v):")
print(az.summary(idata1, var_names=["s_w", "s_v"]).to_string())

# NumPyro built-in model visualization (graph)
# (Outputs a Graphviz Digraph; saving to file for reference)
gv1 = render_model(local_level_model_forecast,
                   model_args=(jnp.asarray(y_all), pred_term),
                   render_distributions=True)
gv1.render(filename="local_level_model_forecast", format="pdf", cleanup=True)
print('Saved model graph: "local_level_model_forecast.pdf"')

# 3.2 Interpolation (missing data)
nuts2 = NUTS(local_level_model_interpolation)
mcmc2 = MCMC(nuts2, num_warmup=num_warmup, num_samples=num_samples, num_chains=num_chains, progress_bar=False)
mcmc2.run(rng_key, y_obs=jnp.asarray(y_obs_only), obs_idx=jnp.asarray(obs_idx), T=T_NA)
mcmc2.print_summary()
#idata2 = az.from_numpyro(
#    posterior=mcmc2.get_samples(group_by_chain=True),
#    sample_stats=mcmc2.get_extra_fields(group_by_chain=True)
#)
idata2 = az.from_numpyro(posterior=mcmc2)
print("\n[Interpolation model] ArviZ summary (s_w, s_v):")
print(az.summary(idata2, var_names=["s_w", "s_v"]).to_string())

gv2 = render_model(local_level_model_interpolation,
                   model_args=(jnp.asarray(y_obs_only), jnp.asarray(obs_idx), T_NA),
                   render_distributions=True)
gv2.render(filename="local_level_model_interpolation", format="pdf", cleanup=True)
print('Saved model graph: "local_level_model_interpolation.pdf"')

# 3.3 Prediction interval (posterior predictive y_pred)
nuts3 = NUTS(local_level_model_pred_interval)
mcmc3 = MCMC(nuts3, num_warmup=num_warmup, num_samples=num_samples, num_chains=num_chains, progress_bar=False)
mcmc3.run(rng_key, y_obs=jnp.asarray(y_obs_only), obs_idx=jnp.asarray(obs_idx), T=T_NA)
mcmc3.print_summary()
#idata3_base = az.from_numpyro(
#    posterior=mcmc3.get_samples(group_by_chain=True),
    # bring posterior predictive separately below for explicit control
    #sample_stats=mcmc3.get_extra_fields(group_by_chain=True)
#)
idata3_base = az.from_numpyro(posterior=mcmc3)
print("\n[Prediction-interval model] ArviZ summary (s_w, s_v):")
print(az.summary(idata3_base, var_names=["s_w", "s_v"]).to_string())

gv3 = render_model(local_level_model_pred_interval,
                   model_args=(jnp.asarray(y_obs_only), jnp.asarray(obs_idx), T_NA),
                   render_distributions=True)
gv3.render(filename="local_level_model_pred_interval", format="pdf", cleanup=True)
print('Saved model graph: "local_level_model_pred_interval.pdf"')


# --------------------------------------------------------------------------------------
# 4) Posterior samples, posterior predictive, and ArviZ InferenceData assembly
#     (without using forbidden args to az.from_numpyro)
# --------------------------------------------------------------------------------------

# 4.1 Forecast model: get mu_pred draws for plotting
samples1 = mcmc1.get_samples()  # aggregated over chains: shape (chains*samples, ...)
mu_pred_draws = np.asarray(samples1["mu_pred"])  # shape (N, T+pred_term)

# 4.2 Interpolation model: get mu draws for plotting
samples2 = mcmc2.get_samples()
mu_draws_interpolation = np.asarray(samples2["mu"])  # shape (N, T_NA)

# 4.3 Prediction-interval model: posterior predictive y_pred + observed data
#     We will NOT pass observed_data to az.from_numpyro (forbidden).
#     Instead, construct two idatas and concatenate.
post3 = mcmc3.get_samples(group_by_chain=True)
extra3 = mcmc3.get_extra_fields(group_by_chain=True)

# Build posterior_predictive dict by running Predictive using aggregated samples,
# then reshape back to (chain, draw, ...)
# Note: 'y_pred' has already been defined in the model; use Predictive to draw it.
total_draws = post3["s_w"].shape[0] * post3["s_w"].shape[1]
flat_post = {k: np.reshape(np.asarray(v), (total_draws,) + tuple(v.shape[2:])) for k, v in post3.items()}
pred3 = Predictive(local_level_model_pred_interval, posterior_samples=flat_post,
                   return_sites=["y_pred", "mu"])(jax.random.PRNGKey(2),
                                                  y_obs=jnp.asarray(y_obs_only),
                                                  obs_idx=jnp.asarray(obs_idx),
                                                  T=T_NA)
# Reshape posterior predictive back to (chains, draws, ...)
n_chains, n_draws = post3["s_w"].shape[:2]
ppc3 = {k: np.reshape(np.asarray(v), (n_chains, n_draws) + tuple(v.shape[1:])) for k, v in pred3.items()}

# InferenceData for posterior + sample_stats + posterior_predictive (no observed_data here)
#idata3_post = az.from_numpyro(
#    posterior=post3,
#    posterior_predictive={"y_pred": ppc3["y_pred"]},
#    sample_stats=extra3
#)
idata3_post = az.from_numpyro(
    posterior=mcmc3,
    posterior_predictive={"y_pred": ppc3["y_pred"]}  # (chains, draws, T) 形状
)
# Observed data added via from_dict (this avoids the forbidden observed_data arg)
idata3_obs = az.from_dict(observed_data={"y": y_NA})
#idata3 = az.concat(idata3_post, idata3_obs)
idata3 = az.from_numpyro(
    posterior=mcmc3,
    posterior_predictive={"y_pred": ppc3["y_pred"]}
)
#print("\n[Prediction-interval model] InferenceData groups:", list(idata3._groups.keys()))
print("\n[Prediction-interval model] InferenceData groups:", idata3.groups())


# --------------------------------------------------------------------------------------
# 5) Visualization (time-series, posterior, PPC) — labels in English
# --------------------------------------------------------------------------------------

# 5.1 Forecast result: latent state forecast (mu_pred)
plot_ts_with_hdi(
    dates=date_plot,
    draws_2d=mu_pred_draws,
    obs_dates=sales_df_all["date"].iloc[:T_all],
    obs_values=y_all,
    title="Forecast result (Local level model)",
    ylabel="Sales",
    xlabel="Date",
    hdi_prob=0.95
)

# Posterior for scale parameters (forecast model) — ArviZ with hdi_prob
az.plot_posterior(idata1, var_names=["s_w", "s_v"], hdi_prob=0.95)
plt.suptitle("Posterior of scale parameters (forecast model)", y=1.02)
plt.tight_layout()
plt.show()

# Optional forest plot (no 'group' argument used)
az.plot_forest(idata1, var_names=["s_w", "s_v"])
plt.title("Forest plot of scale parameters (forecast model)")
plt.tight_layout()
plt.show()

# 5.2 Interpolation result: latent state over full period with observed points
plot_ts_with_hdi(
    dates=sales_df_NA["date"],
    draws_2d=mu_draws_interpolation,
    obs_dates=sales_df_omit_NA["date"],
    obs_values=sales_df_omit_NA["sales"],
    title="Interpolation result (Local level model)",
    ylabel="Sales",
    xlabel="Date",
    hdi_prob=0.95
)

# Posterior for scale parameters (interpolation model)
az.plot_posterior(idata2, var_names=["s_w", "s_v"], hdi_prob=0.95)
plt.suptitle("Posterior of scale parameters (interpolation model)", y=1.02)
plt.tight_layout()
plt.show()

# 5.3 Prediction interval: posterior predictive distribution of y
# PPC (must pass group="posterior")
#az.plot_ppc(idata3, group="posterior")
az.plot_ppc(idata3, group="posterior", data_pairs={"y_obs": "y_pred"})
plt.suptitle("Posterior predictive check (interpolation + prediction interval)", y=1.02)
plt.tight_layout()
plt.show()

# Also show posterior of scale parameters for this model
az.plot_posterior(idata3, var_names=["s_w", "s_v"], hdi_prob=0.95)
plt.suptitle("Posterior of scale parameters (prediction-interval model)", y=1.02)
plt.tight_layout()
plt.show()

print("\nAll done.")
```
